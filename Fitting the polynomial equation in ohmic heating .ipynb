```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df =  pd.read_csv(r"C:\Users\Lenovo\OneDrive - Danmarks Tekniske Universitet\Desktop\machine learning\Food scince and technology\ohmic_heating_conductivity_vs_heating_rate.csv")
print(df.head())
```

```python
# Converting the columns into the row 

X = df["Conductivity (S/m)"].values.reshape(-1,1)
y = df["Heating Rate (°C/s)"].values.reshape(-1,1)

```

```python
# fitting and converting the polynmial form

poly  = PolynomialFeatures(degree = 3)
x2 = poly.fit_transform(X)
```

```python
# Fitting the model 

model  =  LinearRegression()
model.fit(x2,y)
```

```python
#prediction of heating rate
ypredict = model.predict(x2)
```

```python
# finding the 
msey = mean_squared_error(ypredict,y)
r2 = r2_score(y,ypredict)

print("mean_squared_error =",msey)
print("r2_score =",r2)
```

```python
# Visulization of data of conductance vs heating rate 
plt.scatter(X,y, color = "red", label = "Actual points")
plt.plot(X,ypredict,color = "red",label = "Polynomial Fit (degree=3)")
plt.xlabel("Conductance (S/m)",color = "red")
plt.ylabel("Heating Rate (°C/s)",color = "red")
plt.legend()
plt.title("Correlation between the conductnace and heating rate in ohmic heating ",color = "red")
plt.show()
```
